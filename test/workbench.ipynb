```julia
function buildplot(metric)
    plot(1:sim.TIMESTEPS, mean(track["PCA"][metric], 2),
         1:sim.TIMESTEPS, mean(track["clusterfeck"][metric], 2),
         1:sim.TIMESTEPS, mean(track["fixed-variance"][metric], 2),
         1:sim.TIMESTEPS, mean(track["k-means"][metric], 2),
         1:sim.TIMESTEPS, mean(track["hierarchical"][metric], 2))
end

fig = figure();

subplot(221);
buildplot(:MCC)
grid();
ylabel("Matthews corr. coeff.");
legend(["PCA", "clusterfeck", "fixed-variance", "k-means", "hierarchical"],
       bbox_to_anchor=(-0.15, 1.35),
       loc="upper left",
       ncol=4);

subplot(222);
buildplot(:spearman)
ax = gca();
ax[:yaxis][:set_ticks_position]("right");
ax[:set_xscale]("linear");
ylabel("spearman rank corr.");
grid();

subplot(223);
buildplot(:beats)
ax = gca();
ax[:set_xscale]("log");
xlabel("reporting round");
ylabel("beats");
grid();

subplot(224);
buildplot(:liars_bonus)
xlabel("reporting round");
ylabel("liars' bonus");
ax = gca();
ax[:yaxis][:set_ticks_position]("right");
ax[:set_xscale]("log");
grid();

fig[:canvas][:draw]();
```
